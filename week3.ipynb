```python
f=open('/中獎號碼.txt')
text = f.read()
win = text.split()
g=open('/我的發票.txt')
text = g.read()
mine = text.split()
for i, num in enumerate(mine):
  if num in win:
    print('第',i+1,'張發票對中號碼',num,"!")
```
