```python
results = []

hwper = float(input('請輸入作業比重'))
midtermper = float(input('請輸入期中考比重'))
endtermper = float
endtermper = 1-hwper-midtermper


with open('/學生成績2.csv') as infile:
  data = infile.read().split()
  for e in data[1:]:
    stu = e.split(',')
    scores = [int (sc) for sc in stu[1:]]
    exams = scores[5] * midtermper + scores[6] * endtermper
    hw = sum(sorted(scores[0:5])[2:])/3
    final = round(hw*hwper + exams,2)
    print(stu[0], scores, '作業平均', hw, '總分', final)
    stu.append(final)
    results.append(stu)

with open('/新學生成績2.csv', 'w') as outfile:
  outfile.write (data[0]+',總成績\n')
  for stu in results:
    outfile.write(','.join([str(e) for e in stu])+'\n')
```
