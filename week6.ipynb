```python
def factor(x):
  print(x,"的因數有:")
  for i in range(1,x+1):
    if x % i ==0:
      print(i, end=' ')

num = int(input('請輸入正整數:'))
factor(num)
```

```python
def triangle(a,h):
  return 0.5*a*h

a=float(input("輸入三角形的底:"))
h=float(input("輸入三角形的高:"))
print('三角形面積=',triangle(a,h))
```

```python
import math
def area(r):
  a=math.pi*r**2
  return a

def volume(r):
  v=(4.0/3.0)*math.pi*r**3
  return v

r=int(input('輸入半徑='))
print('半徑為','r','的圓面積為:',area(r))
print('半徑為','r','的球體面積為:',volume(r))
```

```python
import csv
from collections import Counter

with open('/中華職棒球員打擊數據.csv') as infile:
  data = list (csv.DictReader(infile))
  results = []
  for e in data:
    if int (e['全壘打']) >=10 and int (e['盜壘']) >=10:
      results.append (e)

counts = Counter([e['姓名'] for e in results])
print('共有', len(counts), '人有此成就')
for r, p in enumerate(counts.most_common (3)):
  name, times = p[0], p[1]
  print('第', r+1, '名:', name, '紀錄達成', times, '次')
  for e in results:
    if e['姓名'] == name:
      print (name, e['年度'], e['全壘打'], e['盜壘'])
```
