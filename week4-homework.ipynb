```python
money = 0

win=open('/中獎號碼.txt').read().split()
mine=open('/我的發票.txt').read().split()

for i, num in enumerate(mine):
  for win_num in win:
    if num[-6:] == win_num[-6:]:
      print ('第', i+1, '張發票對中號碼', num[-6:], '!')
      money=money+10000
    elif num[-5:] == win_num[-5:]:
      print ('第', i+1, '張發票對中號碼', num[-5:], '!')
      money=money+4000
    elif num[-4:] == win_num[-4:]:
      print ('第', i+1, '張發票對中號碼', num[-4:], '!')
      money=money+1000
    elif num[-3:] == win_num[-3:]:
      print ('第', i+1, '張發票對中號碼', num[-3:], '!')
      money=money+200
print('你中了', money, '元!')
```
