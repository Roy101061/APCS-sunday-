```python
import csv
from collections import Counter

a=[];acheck=[];b=[];bcheck=[]

with open('/中華職棒球員打擊數據.csv') as infile:
  data = list (csv.DictReader(infile))

for e in data:
  if int (e['全壘打']) >=20 and int (e['盜壘']) >=20:
    a.append(e['姓名'])
  if int (e['全壘打']) >=20 and float (e['打擊率']) >=0.3:
    b.append(e['姓名'])

print ('年度達到20支全壘打且20次盜壘的選手為:')

for x in a:
  if x not in acheck:
    print(x,end=' ')
    acheck.append(x)


print ('\n年度達到20支全壘打且打擊率超過三成的前三位次數最高選手為:')

C = dict(Counter(b).most_common (3))
C = list(C)
print(C[0], C[1], C[2])
```
